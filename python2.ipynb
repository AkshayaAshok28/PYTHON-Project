```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv("/content/indian_food.csv")
df
```

```python
#🔹 Matplotlib (Basic Plotting)
# 1. Bar chart - number of dishes in each region
df['region'].value_counts().plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Number of Dishes')
plt.title('Number of Dishes per region')
plt.show()
```

```python
# 2. Pie chart - vegetarian vs non-vegetarian
df['diet'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=['green','red'])
plt.title('Vegetarian vs Non-Vegetarian')
plt.show()
```

```python
#3.Histogram - cook_time distribution
df['cook_time'].hist(bins=20,color='skyblue',edgecolor='black')
plt.xlabel('Cook time in minutes')
plt.ylabel('Frequency')
plt.title('Distribution frequency of cook time')
plt.show()
```

```python
#4. Line chart - dishes by prep_time
df['prep_time'].value_counts().plot(kind='line',color='purple',marker='o')
plt.xlabel('Prep time in minutes')
plt.ylabel('Number of Dishes')
plt.title('Dishes by prep time')
plt.show()
```

```python
#🔹 Seaborn (Visualizations)
#1.Countplot - dishes by course
sns.countplot(x='course',data=df,palette='Set2')
plt.title('Dishes by Course')
plt.show()
```

```python
#2.Boxplot - cook_time grouped by diet
sns.boxplot(x='diet',y='cook_time',data=df,palette='coolwarm')
plt.title('Cook time by diet')
plt.show()
```

```python
#3.Violin plot - prep_time across regions
sns.violinplot(x='region',y='prep_time',data=df,palette='coolwarm')
plt.title('Prep time across regions')
plt.show()
```

```python
#4.Sea born Heat map for missing values
sns.heatmap(df.isnull(),cmap='coolwarm',cbar=False)
plt.title("Missing values in Indian food")
plt.show()
```

```python
df=pd.read_csv("/content/indian_food.csv")
df
```

```python
#MERGE AND CONCAT:
# Table 1 - Dish Info
df1=df[['name','diet','course','region']].iloc[:150]
```

```python
# Table 2 - Time and flavor profile:
df2=df[['name','prep_time','cook_time','flavor_profile','state']].iloc[100:]
```

```python
print("Dish info table:")
df1
```

```python
print("Time and flavor profile info:")
df2
```

```python
inner_join=pd.merge(df1,df2,on='name',how='inner')
inner_join
```

```python
left_join=pd.merge(df1,df2,on='name',how='left')
left_join
```

```python
right_join=pd.merge(df1,df2,on='name',how='right')
right_join
```

```python
outer_join=pd.merge(df1,df2,on='name',how='outer')
outer_join
```

```python
#Concat (Stacking Tables)
veg=df[df['diet']=='vegetarian'].head(50)
veg
non_veg=df[df['diet']=='non_vegetarian'].head(50)
non_veg
union=pd.concat([veg,non_veg],axis=0)
union
```
